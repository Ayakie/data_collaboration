```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(palette='Set2', style='white') # context='paper'
seq = sns.color_palette('bone_r', 7)

palette = sns.color_palette('Accent_r')
sns.palplot(palette)
```

# increasing the number of data

```python
df_ndat = pd.read_csv('../save/papers/ndat.csv', index_col=0)
df_ndat
```

```python
# mnist
# mean 
mnist_ndat_cntr_m = df_ndat.loc['mnist_cntr_m']
mnist_ndat_ind_m = df_ndat.loc['mnist_ind_m']

mnist_ndat_dc_mlp_m = df_ndat.loc['mnist_dc_mlp_m']
mnist_ndat_dc_svm_m = df_ndat.loc['mnist_dc_svm_m']
mnist_ndat_dc_knn_m = df_ndat.loc['mnist_dc_knn_m']

mnist_ndat_fed_m = df_ndat.loc['mnist_fed_m']

# std
mnist_ndat_cntr_s = df_ndat.loc['mnist_cntr_s']
mnist_ndat_ind_s = df_ndat.loc['mnist_ind_s']

mnist_ndat_dc_mlp_s = df_ndat.loc['mnist_dc_mlp_s']
mnist_ndat_dc_svm_s = df_ndat.loc['mnist_dc_svm_s']
mnist_ndat_dc_knn_s = df_ndat.loc['mnist_dc_knn_s']

mnist_ndat_fed_s = df_ndat.loc['mnist_fed_s']
```

```python
# fashon-mnist
# mean 
fmnist_ndat_cntr_m = df_ndat.loc['fmnist_cntr_m']
fmnist_ndat_ind_m = df_ndat.loc['fmnist_ind_m']

fmnist_ndat_dc_mlp_m = df_ndat.loc['fmnist_dc_mlp_m']
fmnist_ndat_dc_svm_m = df_ndat.loc['fmnist_dc_svm_m']
fmnist_ndat_dc_knn_m = df_ndat.loc['fmnist_dc_knn_m']

fmnist_ndat_fed_m = df_ndat.loc['fmnist_fed_m']
fmnist_ndat_fed_3E_m = df_ndat.loc['fmnist_fed_3e_m']

# std
fmnist_ndat_cntr_s = df_ndat.loc['fmnist_cntr_s']
fmnist_ndat_ind_s = df_ndat.loc['fmnist_ind_s']

fmnist_ndat_dc_mlp_s = df_ndat.loc['fmnist_dc_mlp_s']
fmnist_ndat_dc_svm_s = df_ndat.loc['fmnist_dc_svm_s']
fmnist_ndat_dc_knn_s = df_ndat.loc['fmnist_dc_knn_s']

fmnist_ndat_fed_s = df_ndat.loc['fmnist_fed_s']
fmnist_ndat_fed_3e_s = df_ndat.loc['fmnist_fed_3e_s']
```

```python
plt.figure(figsize=(12,6))
xval = df_ndat.columns
plt.plot(xval, df_ndat.loc['mnist_cntr_m', :], color='grey', marker='.', linestyle='--', label='Centralized data analysis')
plt.plot(xval, mnist_ndat_ind_m, color='grey', marker='.', linestyle='--', label='Individual data analysis')
plt.plot(xval, mnist_ndat_dc_knn_m, marker='.', linestyle=':', label='DC-knn, knn=5')
plt.plot(xval, mnist_ndat_dc_svm_m, marker='.', linestyle=':', label='DC-svm, C=10')
plt.plot(xval, mnist_ndat_dc_mlp_m, marker='.', color=palette[1], label='DC-mlp, adam')
plt.plot(xval, mnist_ndat_fed_m, marker='.', color=palette[2], label='Federated Learning, 1E32B')
plt.ylabel('Accuracy')
plt.xlabel('Number of data per user (5 users)')
locs, labels = plt.xticks()
plt.title("Accuracy (Mnist)")
plt.legend()
```

```python
plt.figure(figsize=(12,6))
plt.plot(xval, fmnist_ndat_cntr_m, color='grey', marker='.', linestyle='--', label='Centralized data analysis')
plt.plot(xval, fmnist_ndat_ind_m, color='grey', marker='.', linestyle='--', label='Individual data analysis')
plt.plot(xval, fmnist_ndat_dc_knn_m, marker='.', linestyle=':', label='DC-knn, knn=5')
plt.plot(xval, fmnist_ndat_dc_svm_m, marker='.', linestyle=':', label='DC-svm, C=10')
plt.plot(xval, fmnist_ndat_dc_mlp_m, marker='.', color=palette[1], label='DC-mlp, adam')
plt.plot(xval, fmnist_ndat_fed_m, marker='.', linestyle='-.', color=palette[2], label='Federated Learning, 1E32B')
plt.plot(xval, fmnist_ndat_fed_3E_m, marker='.', color=palette[2], label='Federated Learning, 3E32B')
plt.ylabel('Accuracy')
plt.xlabel('Number of data per user (5 users)')
locs, labels = plt.xticks()
plt.xticks(xval)
plt.title("Accuracy (Fashion-mnist)")
plt.legend()
```
